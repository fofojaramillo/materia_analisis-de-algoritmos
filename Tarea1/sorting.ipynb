Realizar la graficaci´on de las siguientes funciones: f(n) =
log(n), f(n) = n, f(n) = n · log(n), f(n) = n
2
, f(n) = n
3
.
Todas las curvas deben ser colocadas en la misma gr´afica

```python
n = np.linspace(0.01, 2, 100)
f_log=np.log10(n)
f=n
f_nlog=n*np.log10(n)
f_2=n**2
f_3=n**3

plt.figure(figsize=(7, 8))
plt.xlabel('n')
plt.ylabel('f(n)')
plt.plot(n, f_log, 'r', label='log(n)')
plt.plot(n, f, 'b', label='n')
plt.plot(n, f_nlog, 'g', label='nlog(n)')
plt.plot(n, f_2, 'y', label='n^2')
plt.plot(n, f_3, 'm', label='n^3')
plt.ylim([0, 8])
plt.legend()
plt.savefig('grafica-funciones.png')
```
